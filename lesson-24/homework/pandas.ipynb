```python
##Homework
##Assignment 1
##1. Group the data by the `Category` column
#  and calculate the following aggregate statistics for each category:
   #- Total quantity sold.
#    - Average price per unit.
#    - Maximum quantity sold in a single transaction.
```

```python
import pandas as pd

sales_data = pd.read_csv('sales_data.csv')
category=(
    sales_data
    .groupby('Category')
    .agg(
        Total_Quantity_Sold=('Quantity', 'sum'),
        Avg_Price=('Price', 'mean'),
        Max_Quantity=('Quantity', 'max')
    )
    .round(2)
)
print(category)
```

```python
sales_data=pd.read_csv('sales_data.csv')
sales_data
```

```python
##2. Identify the top-selling product in each category based on the total quantity sold.
```

```python
top_products=(
    sales_data
    .groupby(['Category', 'Product'])['Quantity']
    .sum()
    .reset_index()
    .sort_values(['Category', 'Quantity'], ascending=[True, False])
    .drop_duplicates(subset=['Category'], keep='first')
)
print(top_products)
```

```python
##3. Find the date on which the highest total sales (quantity * price) occurred.
```

```python
sales_data['Total_Sales'] = sales_data['Quantity'] * sales_data['Price']
highest_total_sale=(
    sales_data
    .groupby('Date')['Total_Sales']
    .sum()
    .idxmax()
)

print(highest_total_sale)
```

```python
##**Homework Assignment 2: Examining Customer Orders**

#You have a dataset (`task\customer_orders.csv`) containing information about customer orders. The dataset has the following columns:

# - `OrderID`: Unique identifier for each order.
# - `CustomerID`: Unique identifier for each customer.
# - `Product`: Name of the product ordered.
# - `Quantity`: Number of units ordered.
# - `Price`: Price per unit.

# **Tasks:**

# 1. Group the data by `CustomerID` and filter out customers who have made less than 20 orders.
# 2. Identify customers who have ordered products with an average price per unit greater than $120.
# 3. Find the total quantity and total price for each product ordered, and filter out products that have a total quantity less than 5 units.
```

```python
orders=pd.read_csv('customer_orders.csv')
orders
```

```python
# 1. Group the data by `CustomerID` and filter out customers who have made less than 20 orders.
```

```python
less_20_orders=(
    orders
    .groupby('CustomerID')['OrderID']
    .count()
    .reset_index()
    .query('OrderID >=20 ')

)

filtered_customers = orders[orders['CustomerID'].isin(less_20_orders['CustomerID'])]

print(filtered_customers)
```

```python
# 2. Identify customers who have ordered products with an average price per unit greater than $120.
```

```python
customers= (
    orders
    .groupby('CustomerID')['Price']
    .mean()
    .reset_index()
    .query('Price>120')
)
print(customers)
```

```python
# 3. Find the total quantity and total price for each product ordered, 
# and filter out products that have a total quantity less than 5 units.
```

```python
product=(
    orders
    .groupby('Product')
    .agg(
        Total_Quantity=('Quantity', 'sum'),
        Total_Price=('Quantity','sum' )
    )
    .reset_index()
    .query('Total_Quantity >=5')
)
print(product)
```
