```python
# survival count for each passenger class

import matplotlib.pyplot as plt
import seaborn as sns
titanic = sns.load_dataset('titanic')
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))
sns.countplot(data=titanic,x="class", hue="survived", palette="coolwarm")

plt.xlabel("Passenger class")
plt.ylabel("Counting")
plt.title("Survival Count of Passenger Class")
plt.show()

```

```python
#distribution of fares paid by passengers in different classes 

sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))

sns.boxplot(data=titanic, x="class", y="fare", hue="class", palette="viridis", legend=False)

plt.xlabel("Passenger Class")
plt.ylabel("Fare price($)")
plt.title("Fare distrubution by Passenger Class")

plt.show()
```

```python
#ages are distributed among Titanic passengers
plt.figure(figsize=(10, 6))
sns.histplot(titanic["age"], bins=30, kde=True, color='g')
plt.title("Age distribution of Titanic passenger")
plt.xlabel("Age")
plt.ylabel("Counting")
plt.show()
```

```python
sns.pairplot(titanic, hue="survived", diag_kind="kde", palette="viridis")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=titanic, x="age", y="fare", hue="survived", palette="coolwarm", alpha=0.7)
plt.title("Age and Fare (Colored by Survival)")
plt.xlabel("Age")
plt.ylabel("Fare Price ($)")
plt.show()
```
